<span style="color: Gold"> VIT 모델</span> (가상환경 3.10이 가장 안정적)

<span style="font-size:12px;">

- 이미지를 잘게 쪼개서 “문장처럼” 보고, 트랜스포머로 이해하는 모델
- 패치 → 토큰 → 어텐션이 전부
    - 작동원리
        - 이미지를 패치로 나누기 ex)224x224 --> 16x16 196개 패치
        - 패치를 벡터로 변환
        - 위치정보를 추가 : 각 패치가 이미지의 어디에 위치했는지 정보를 추가
        - Transformer 처리 - 학습
        - 분류
- CNN과의 차이점
    - CNN은 특정 이미지의 부분을 보는 관점, VIT는 이미지 전체를 한번에 볼 수 있음

<span style="color: Gold"> ViT 내부 구조

<span style="font-size:12px;">

```text

이미지 (H x W x C)
   ↓
Patch 분할 (예: 16x16)
   ↓
Flatten → Linear Embedding
   ↓
[CLS] 토큰 추가
   ↓
Position Embedding 추가
   ↓
Transformer Encoder (Self-Attention)
   ↓
[CLS] 출력 → 분류
```
