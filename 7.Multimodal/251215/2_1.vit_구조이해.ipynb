<span style="color: Gold"> 1. íŒ¨ì¹˜ë¶„í• 

````python
def patch_embedding():
    '''ì´ë¯¸ì§€ë¥¼ íŒ¨ì¹˜ë¡œ ë¶„í• í•˜ëŠ” ê³¼ì •(patch embedding)'''
    # ì„¤ì •
    image_size = 224
    patch_size = 16
    channels = 3
    embedding_dim = 768

    # íŒ¨ì¹˜ìˆ˜ ê³„ì‚°
    num_patchs = (image_size // patch_size) **2
    print(f' ì´ë¯¸ì§€ í¬ê¸° : {image_size} x {image_size }')
    print(f' íŒ¨ì¹˜ í¬ê¸° : {patch_size} x {patch_size }')
    print(f' ì±„ë„ ìˆ˜ : {channels}')
    print(f' íŒ¨ì¹˜ ìˆ˜ : {image_size // patch_size} x{image_size // patch_size}')

    # ë”ë¯¸ ì´ë¯¸ì§€ ìƒì„±
    dummy_image = torch.randn(1,channels, image_size, image_size)
    print(f' ë”ë¯¸ ì´ë¯¸ì§€ ìƒì„±')
    print(f' ì…ë ¥ ì´ë¯¸ì§€ shape : {dummy_image.shape}') # [1,3,224,224]

    # íŒ¨ì¹˜ë¶„í• (conv2 ì‚¬ìš©)
    # conv2 stride = patch_size ê²¹ì¹˜ì§€ ì•ŠëŠ” íŒ¨ì¹˜ ì¶”ì¶œ / # stride = ì»¤ë„ì´ í•œ ë²ˆì— ëª‡ ì¹¸ì”© ì´ë™í•˜ëŠëƒ
    patch_embed = nn.Conv2d(in_channels=channels, out_channels=embedding_dim, kernel_size=patch_size, stride=patch_size) # ì…ë ¥í•˜ëŠ” ì±„ë„ì˜ ìˆ˜, ì¶œë ¥í•˜ëŠ” ì±„ë„ì˜ ìˆ˜, kernel_size -> ëª‡ê°œì”© ë„ì¥ì„ ì°ëŠëƒ, stride -> ì´ë™ ê°„ê²©

    # íŒ¨ì¹˜ ì„ë² ë”© ì ìš©
    patches = patch_embed(dummy_image)
    print(f' \n íŒ¨ì¹˜ ì„ë² ë”© í›„')
    print(f' conv2d ì¶œë ¥ shape : {patches.shape}') # [1, 768, 14, 14]

    # Flatten : (ë°°ì¹˜ì‚¬ì´ì¦ˆ(B), ì„ë² ë”© ì°¨ì´(D), ì´ë¯¸ì§€ ë†’ì´(H), ì´ë¯¸ì§€ ë„“ì´(W)) -> (B, 196(HxW), D) -> (1, 196, 768)
    patches_flat = patches.flatten(2).transpose(1,2)
    print(f' Flatten í›„ shape : {patches_flat.shape}') # [1, 196, 768]

    # ê° íŒ¨ì¹˜ëŠ” 768ì°¨ì› ë²¡í„°
    print(f' \n íŒ¨ì¹˜ ìˆ˜  : {patches_flat.shape[1]}')
    print(f' ê° íŒ¨ì¹˜ì˜ ì„ë² ë”© ì°¨ì› ìˆ˜  : {patches_flat.shape[2]}')
    return patches_flat

if __name__ == '__main__':
    patch_embedding()
````

<span style="color: Gold"> 2. ìœ„ì¹˜ ì„ë² ë”©ì˜ ì—­í•

````python
def positional_embedding():
    '''ìœ„ì¹˜ ì„ë² ë”©'''
    num_patches = 196
    embedding_dim = 768

    # ìœ„ì¹˜ ì„ë² ë”© ìƒì„±  
    # ì´ í…ì„œëŠ” í•™ìŠµëŒ€ìƒ Optimizer ì— ì˜í•´ ì—…ë°ì´íŠ¸
    position_embedding = nn.Parameter(torch.randn(1, num_patches+1, embedding_dim)) # num_patches+1 -> +1ì„ í•˜ëŠ” ì´ìœ ëŠ” cls í† í°ì„ ì¶”ê°€í•˜ê¸° ìœ„í•´ì„œ
    print(f' ìœ„ì¹˜ ì„ë² ë”© shape : {position_embedding.shape}')
    print(f' ì´ ìœ„ì¹˜ ìˆ˜ : {num_patches+1} (íŒ¨ì¹˜ 196 + cls í† í° 1ê°œ)')

    # ë°°ì¹˜ ì°¨ì› ì œê±° -> ê° ìœ„ì¹˜ë¥¼ í•˜ë‚˜ì˜ ë²¡í„°ë¡œ ë‹¤ë£¨ê¸° ìœ„í•´ ë°°ì¹˜ í¬ê¸°ê°€ 1ì¸ í˜•íƒœëŠ” ë¶„ì„ ì‹œ ë¶ˆí•„ìš”
    pos_emb = position_embedding.squeeze(0) #squeeze(0)-> ì²«ë²ˆì§¸(ë°°ì¹˜í¬ê¸°)ë¥¼ ìˆ¨ê¸´ë‹¤

    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°ì„ ìœ„í•´ì„œ ì •ê·œí™”ëŠ” í•„ìˆ˜ -> ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ í•˜ëŠ” ì´ìœ ëŠ” ì˜ë¯¸ì˜ ë°©í–¥ë§Œ ë¹„êµí•˜ë ¤ê³  ì‚¬ìš©
    # ìœ„ì¹˜ ì •ë³´
    # â†’ positional embeddingìœ¼ë¡œ í•´ê²°
    # ì˜ë¯¸ ê´€ê³„
    # â†’ attention + ì½”ì‚¬ì¸ ìœ ì‚¬ë„(ì •ê·œí™” ë‚´ì )ë¡œ í•´ê²°
    pos_emb_norm = pos_emb/pos_emb.norm(dim=1,keepdim=True) # ë‹¨ìœ„ë²¡í„°ë¡œ ë‚˜ëˆ”
    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ í–‰ë ¬
    similarity = torch.mm(pos_emb_norm,pos_emb_norm.t())    # [197,197]
    # ì§€ê¸ˆì€ í•™ìŠµì „ì´ë¼ì„œ ëœë¤ì´ì§€ë§Œ í•™ìŠµ í›„ì—ëŠ” ì¸ì ‘ íŒ¨ì¹˜ê¹Œì§€ ìœ ì‚¬í•´ì§
    return position_embedding

if __name__ == '__main__':
    positional_embedding()
````

<span style="color: Gold"> 3. CLS í† í°

````python
def cls_token():
    embedding_dim = 768
    num_paches = 196
    batch_size = 2
    
    # cls í† í° ìƒì„±
    cls_token = nn.Parameter(torch.rand(1,1,embedding_dim)) # í•™ìŠµì„ í†µí•´ ì´ë¯¸ì§€ë¥¼ ìš”ì•½í•˜ëŠ” ë²¡í„°ë¡œ ì§„í™”
    print(f' CLS í† í° shape : {cls_token.shape}')   # [1,1,768]
    # ë°°ì¹˜ í¬ê¸°ì— ë§ê²Œ CLS í† í° í™•ì¥
    cls_tokens = cls_token.expand(batch_size,-1,-1) # ë°°ì¹˜ë§ˆë‹¤ ë™ì¼í•œ cls í† í°ì´ í•„ìš”, expandëŠ” viewë§Œ í™•ì¥
    print(f' ë°°ì¹˜ í™•ì¥ shape : {cls_tokens.shape}')   # [2,1,768]
    # íŒ¨ì¹˜ ì„ë² ë”© ìƒì„±
    patch_embeddings = torch.randn(batch_size, num_paches, embedding_dim)
    print(f' íŒ¨ì¹˜ ì„ë² ë”© shape : {patch_embeddings.shape}')
    # CLS í† í°ì„ ì‹œí€€ìŠ¤ ë§¨ ì•ì— ì¶”ê°€
    embeddings = torch.cat([cls_tokens,patch_embeddings], dim=1)    # dim=1 -> í† í° ì°¨ì› / [cls, patch1 ,... patch 196]
    print(f'ê²°í•©í›„shape : {embeddings.shape}') # [2,197,768]
    return embeddings

if __name__ == '__main__':
    cls_token()
````

<span style="color: Gold"> 4. self-attention

````python
def self_attention():
    '''self_aattention ë©”ì»¤ë‹ˆì¦˜
    Mult-Head self-attentionì˜ êµ¬ì„±ìš”ì†Œ
    Q(query)
    K(key)
    V(value)
    '''
    embedding_dim = 768
    num_heads = 12
    head_dim = embedding_dim//num_heads # ê° í—¤ë“œê°€ ë‹´ë‹¹í•˜ëŠ”ì°¨ì›
    selq_len = 197 # 196íŒ¨ì¹˜ +1 cls
    batch_size = 1  # ìƒ˜í”Œ1ê°œ
    # ë”ë¯¸ë°ì´í„° ìƒì„±
    x = torch.randn(batch_size,selq_len,embedding_dim) # [1,197,768]
    # QKV ì„ í˜• ë ˆì´ì–´
    qkv_proj = nn.Linear(embedding_dim,embedding_dim*3) # í•˜ë‚˜ì˜ Linear ì—°ì‚°ìœ¼ë¡œ QKV ë™ì‹œì— ìƒì„±
    # QKV ê³„ì‚°
    qkv = qkv_proj(x)
    print(f' QKV shape : {qkv.shape}')
    # Q K V ë¶„ë¦¬
    qkv = qkv.reshape(batch_size, selq_len,3, num_heads, head_dim) # [B, N, 3, heads, head_dim]
    qkv = qkv.permute(2,0,3,1,4) # [3,B,heads,N,head_dim]
    q, k , v = qkv[0], qkv[1], qkv[2]
    print(f' Q shape : {q.shape}')
    print(f' K shape : {k.shape}')
    print(f' V shape : {v.shape}')


# Attention Score ê³„ì‚°
    scale = head_dim ** -0.5
    attn = (q @ k.transpose(-2, -1)) * scale
    print(f"\n[Attention Score]")
    print(f"  Attention shape: {attn.shape}")  # [1, 12, 197, 197]
    
    # Softmax
    attn = attn.softmax(dim=-1)
    print(f"  Softmax í›„ í•©ê³„ (ê° í–‰): {attn[0, 0, 0].sum().item():.4f}")  # 1.0
    
    # Valueì™€ ê³±í•˜ê¸°
    out = attn @ v
    print(f"\n[Attention ì ìš©]")
    print(f"  ì¶œë ¥ shape: {out.shape}")  # [1, 12, 197, 64]
    
    # í—¤ë“œ ê²°í•©
    out = out.transpose(1, 2).reshape(batch_size, selq_len, embedding_dim)
    print(f"  í—¤ë“œ ê²°í•© í›„: {out.shape}")  # [1, 197, 768]
    
    return attn


if __name__ == '__main__':
    self_attention()
    
````

<span style="color: Gold"> 5. Multi-Layer Perceptron </span>   
- Attentionì´ ëª¨ì€ ì •ë³´ë¥¼ ë‹¤ì‹œ ê°€ê³µí•˜ëŠ” ë‹¨ê³„
- Attentionì€ â€œì–´ë””ë¥¼ ë³¼ì§€â€ ê²°ì •í•˜ê³ , MLPëŠ” â€œë³¸ ê±¸ ì–´ë–»ê²Œ í•´ì„í• ì§€â€ ê²°ì •

````python
def mlp():
    '''mlp ë¸”ëŸ­'''
    embedding_dim = 768
    mlp_dim = embedding_dim*4 # ì¼ë°˜ì ìœ¼ë¡œ 4ë°° í™•ì¥
    print(f' ì…ë ¥/ ì¶œë ¥ ì°¨ì› : {embedding_dim}')
    print(f'íˆë“ ì°¨ì› : {mlp_dim}')
    # MLP ë¸”ëŸ­ ì •ì˜
    mlp = nn.Sequential(
        nn.Linear(embedding_dim,mlp_dim),
        nn.GELU(), # ë‹¨ìˆœì„ í˜•ì„ ë¹„ì„ í˜•ìœ¼ë¡œ ë³€í˜•
        nn.Linear(mlp_dim,embedding_dim)
    )
    
     # íŒŒë¼ë¯¸í„° ìˆ˜ ê³„ì‚°
    total_params = sum(p.numel() for p in mlp.parameters())
    print(f"  MLP íŒŒë¼ë¯¸í„° ìˆ˜: {total_params:,}")
    
    # ë”ë¯¸ ì…ë ¥
    x = torch.randn(1, 197, embedding_dim)
    print(f"\n[ì…ë ¥/ì¶œë ¥]")
    print(f"  ì…ë ¥ shape: {x.shape}")
    
    # MLP ì ìš©
    out = mlp(x)
    print(f"  ì¶œë ¥ shape: {out.shape}")
    
    print(f"\n[GELU í™œì„±í™” í•¨ìˆ˜]")
    print(f"  GELU(x) = x * Phi(x)")
    print(f"  ReLUë³´ë‹¤ ë¶€ë“œëŸ½ê³ , ìŒìˆ˜ ì…ë ¥ì—ë„ ì‘ì€ ê°’ ì¶œë ¥")
    
    return out
````

<span style="color: Gold"> 6. transformer encoder block êµ¬ì¡°

````python
def transformer_block():
    '''Transformer Encoder ë¸”ëŸ­'''
    class TransformerBlock(nn.Module):
        """ê°„ë‹¨í•œ Transformer Block êµ¬í˜„"""
        def __init__(self, dim=768, num_heads=12, mlp_ratio=4.0):
            super().__init__()
            self.norm1 = nn.LayerNorm(dim)
            self.attn = nn.MultiheadAttention(dim, num_heads, batch_first=True)
            self.norm2 = nn.LayerNorm(dim)
            self.mlp = nn.Sequential(
                nn.Linear(dim, int(dim * mlp_ratio)),
                nn.GELU(),
                nn.Linear(int(dim * mlp_ratio), dim),
            )
        
        def forward(self, x):
            # Pre-norm êµ¬ì¡°
            # Attention with residual
            x = x + self.attn(self.norm1(x), self.norm1(x), self.norm1(x))[0]
            # MLP with residual
            x = x + self.mlp(self.norm2(x))
            return x
    
    # ë¸”ë¡ ìƒì„±
    block = TransformerBlock()
    
    # íŒŒë¼ë¯¸í„° ìˆ˜
    total_params = sum(p.numel() for p in block.parameters())
    print(f"\n[Transformer Block êµ¬ì„±]")
    print(f"  1. Layer Normalization")
    print(f"  2. Multi-Head Self-Attention")
    print(f"  3. Residual Connection")
    print(f"  4. Layer Normalization")
    print(f"  5. MLP (Feed-Forward)")
    print(f"  6. Residual Connection")
    print(f"\n  ë¸”ë¡ë‹¹ íŒŒë¼ë¯¸í„° ìˆ˜: {total_params:,}")
    
    # ë”ë¯¸ ì…ë ¥
    x = torch.randn(1, 197, 768)
    print(f"\n[ì…ë ¥/ì¶œë ¥]")
    print(f"  ì…ë ¥ shape: {x.shape}")
    
    out = block(x)
    print(f"  ì¶œë ¥ shape: {out.shape}")
    
    # ViT-BaseëŠ” 12ê°œ ë¸”ë¡
    print(f"\n[ViT-Base ì „ì²´ íŒŒë¼ë¯¸í„°]")
    print(f"  12ê°œ ë¸”ë¡ íŒŒë¼ë¯¸í„°: {total_params * 12:,}")
    
    return block
````
