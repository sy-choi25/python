<span style="color: Gold"> 서버 요청 절차

```text
브라우저 / 클라이언트
        |
        |  GET /hello 요청
        v
---------------------
|     FastAPI 서버   |
---------------------
        |
        |  라우터 매칭
        v
   say_hello() 실행
        |
        v
 {'message': 'Hello world'} → JSON 응답
 ```

 ---

보통 파이썬 함수는 딕셔너리를 리턴하면 그냥 딕셔너리다.  
하지만 FastAPI에서는 다르다

```text
Python dict
   ↓
JSON으로 변환
   ↓
HTTP Response로 전송
```
---

<span style="color: Gold"> GET / POST 차이
- GET은 읽기 전용 질문/ ex)목록조회
- POST는 행동 요청/ ex)회원가입
---

````python
# 데코레이터 @app.get('/hello') --- 라우터: 경로를 지정해주는 것 
# http://localhost:8000/hello -> 원래 로컬 경로에 "/hello"  -> 서버내의 hello 함수를 실행
# app.get('/hello') -> 누구를 실행할지 라우터를 보고
# 함수

@app.get('/hello')
def say_hello():
    return {'message' : 'Hello world'}
````

````python
app = FastAPI(
    title = 'FAST API',
    description = '기본기능 확인',
    version = '0.0.1'
)
````

<span style="color: Gold">
경로 파라미터 vs 쿼리 파라미터

````python
# 경로 파라메터
# http://localhost:8000/hello/홍길동
@app.get('/hello/{name}')
def say_hello(name:str):
    return {'message' : f'Hello {name}'}        # 자동으로 json 형태로 변환
# => Hello 홍길동    이 출력

# 쿼리 파라미터
# http://localhost:8000/greet?name=홍길동;age=35  / 물음표 앞까지는 경로 뒤에는 값에 따라 입력된다
@app.get('/greet')  
def greet(name:str, age:int):
    return {'message' : f'Hello {name}'}
````
