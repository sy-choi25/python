```python
from transformers import PreTrainedTokenizerFast
tokenizer = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2",   # skt/kogpt2-base-v2 -> ÌïúÍµ≠Ïñ¥ GPT-2 Î™®Îç∏
bos_token='</s>', eos_token='</s>', unk_token='<unk>',
pad_token='<pad>', mask_token='<mask>')
tokenizer.tokenize("ÏïàÎÖïÌïòÏÑ∏Ïöî. ÌïúÍµ≠Ïñ¥ GPT-2 ÏûÖÎãàÎã§.üò§:)l^o")

# bos_token Î¨∏Ïû• ÏãúÏûë
# eos_token Î¨∏Ïû• ÎÅù
# unk_token ÏÇ¨Ï†ÑÏóê ÏóÜÎäî Îã®Ïñ¥
# pad_token Ìå®Îî©
# mask_token ÎßàÏä§ÌÇπÏö©
```

```python
import torch
from transformers import GPT2LMHeadModel

model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
text = 'Í∑ºÏú°Ïù¥ Ïª§ÏßÄÍ∏∞ ÏúÑÌï¥ÏÑúÎäî'
input_ids = tokenizer.encode(text, return_tensors='pt')
gen_ids = model.generate(input_ids,
                           max_length=128,
                           repetition_penalty=2.0,
                           pad_token_id=tokenizer.pad_token_id,
                           eos_token_id=tokenizer.eos_token_id,
                           bos_token_id=tokenizer.bos_token_id,
                           use_cache=True)
generated = tokenizer.decode(gen_ids[0])
print(generated)
```

```python
# ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†∏ Î™®Îç∏ÏùÑ Î°úÎìú
import torch
from transformers import GPT2LMHeadModel
from transformers import PreTrainedTokenizerFast

tokenizer = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2",
bos_token='</s>', eos_token='</s>', unk_token='<unk>',
pad_token='<pad>', mask_token='<mask>')

model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
model.eval()
# ÌîÑÎ°¨ÌîÑÌä∏ ÏóîÏßÄÎãàÏñ¥ÎßÅ
prompt = f'{text}\n\ÌïúÏ§Ñ ÏöîÏïΩ:'
# Ïù∏ÏΩîÎî©
input_ids = tokenizer.encode(prompt, return_tensors='pt')
# Î™®Îç∏ Ï∂îÎ°†(ÏÉùÏÑ±)
with torch.no_grad():
  gen_ids = model.generate(input_ids,
                          #  max_length=128,
                           max_new_tokens = 100,
                           repetition_penalty=2.0, # Í∞ôÏùÄ ÎßêÏùÑ Î∞òÎ≥µÌïòÏßÄ ÏïäÎèÑÎ°ù Î≤åÏ†ê
                           pad_token_id=tokenizer.pad_token_id,
                           eos_token_id=tokenizer.eos_token_id,
                           bos_token_id=tokenizer.bos_token_id,
                           use_cache=True,
                           do_sample=True, # ÏûêÏó∞Ïä§Îü¨Ïö¥ Î¨∏Ïû•
                           temperature=0.7, # Ï∞ΩÏùòÏÑ± Ï°∞Ï†ï(ÎÇÆÏùÑÏàòÎ°ù Î≥¥ÏàòÏ†Å, ÎÜíÏùÑÏàòÎ°ù ÌôòÍ∞ÅÍ∞ÄÎä•ÏÑ± ÎÜíÏùå)
                           top_k=50, # ÌôïÎ•† Ï†ÅÏúºÎ°ú ÏÉÅÏúÑ 50Í∞ú Îã®Ïñ¥Ï§ëÏóêÏÑú ÏÑ†ÌÉù
                           )
  generated = tokenizer.decode(gen_ids[0],skip_special_tokens=True)
  # ÌîÑÎ°¨ÌîÑÌä∏ Ïù¥ÌõÑÏóê ÏÉùÏÑ±Îêú ÌÖçÏä§Ìä∏Îßå Ï∂îÏ∂ú
  summary = generated[len(prompt):].strip()
  print(summary)

```
